```python
def filter(level): 

    start = level[0]
    rest = level[1:]
    if rest == []: 
        return 1 
    if start == rest[0] or abs(start-rest[0]>3) or (start< rest[0] and rest[0] > rest[1]) or (start > rest[0] and rest[0] > rest[1]):
        return 0
    else:
        filter(rest)
    return 1 

```
